# Viszualization

```python
# Train.csv
def timeplot(df):
    for col in df.columns.tolist():
        fig = plt.figure(figsize=(16,4))
        plt.plot(df.index, df[col].values)
        plt.title(col)
        plt.show()
        
timeplot(df_train)
```

```python
# pie plot
def pie_plot(df, title):
    trace0 = go.Bar(
        x = df['Scored'].value_counts().index,
        y = df['Scored'].value_counts().values
        )

    trace1 = go.Pie(
        labels = df['Scored'].value_counts().index,
        values = df['Scored'].value_counts().values,
        domain = {'x':[0.55,1]})

    data = [trace0, trace1]
    layout = go.Layout(
        title =  title + ' Distribution for target data',
        xaxis = dict(domain = [0,.50]))

    fig = go.Figure(data = data, layout = layout)
    py.iplot(fig)
```

```python
pie_plot(df_train[df_train['ai1']>=29.9], 'ai1')
```

```python
pie_plot(df_train[df_train['ai3']>=29.9], 'ai3')
```

```python
# predict
timeplot(df_predict)
```

# Correlation

```python
def corr_plot(df, title='Dataset'):

    fig, ax = plt.subplots(1,1, figsize = (15,6))
    hm = sns.heatmap(df.iloc[:,:].corr(),
                    ax = ax,
                    cmap = 'coolwarm',
                    annot = True,
                    fmt = '.2f',
                    linewidths = 0.05)

    fig.subplots_adjust(top=0.93)
    fig.suptitle(title, 
                  fontsize=14, 
                  fontweight='bold')
    
corr_plot(df_train, title= 'train dataset')
```

```python
corr_plot(df_predict, title= 'Predict dataset')
```

```python
# Only train.csv
plt.figure(figsize=(26, 16))
for i, col in enumerate(df_train.columns):
    ax = plt.subplot(4, 3, i + 1)
    sns.distplot(df_train[col], bins=100, label='train')
    ax.legend()  
```

```python
# both train.csv and predict.csv
plt.figure(figsize=(26, 16))
for i, col in enumerate(df_predict.columns):
    ax = plt.subplot(3, 3, i + 1)
    sns.distplot(df_train[col].values, bins=100, label='train')
    sns.distplot(df_predict[col].values, bins=100, label='predict')
    ax.legend()  
```

# Checking Scored (Target) values  

```python
target = df_train['Scored'].value_counts().reset_index().rename(columns = {'index' : 'target'})
target
```

```python
trace0 = go.Bar(
    x = df_train['Scored'].value_counts().index,
    y = df_train['Scored'].value_counts().values
    )

trace1 = go.Pie(
    labels = df_train['Scored'].value_counts().index,
    values = df_train['Scored'].value_counts().values,
    domain = {'x':[0.55,1]})

data = [trace0, trace1]
layout = go.Layout(
    title = 'Frequency Distribution for surface/target data',
    xaxis = dict(domain = [0,.50]))

fig = go.Figure(data = data, layout = layout)
py.iplot(fig)
```

```python
normal  = df_train[df_train["Scored"] == 1].copy()
anomaly= df_train[df_train["Scored"] == -1].copy()
```

```python
# normal vs abnormal
plt.figure(figsize=(26, 16))
for i, col in enumerate(df_train.columns):
    ax = plt.subplot(4, 3, i + 1)
    sns.distplot(normal[col].values, bins=100, label='normal')
    sns.distplot(anomaly[col].values, bins=100, label='anomaly')
    ax.legend()  
    ax.set_title(col)
```
